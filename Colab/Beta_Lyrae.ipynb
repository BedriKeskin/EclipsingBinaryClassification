```python
if __name__ == '__main__':
    import phoebe
    from phoebe import u
    import numpy as np
    import matplotlib.pyplot as plt
    import os
    from google.colab import drive

logger = phoebe.logger()

count = 1

drive.mount('/content/gdrive')
cwd = '/content/gdrive/MyDrive/Doktora/Python'

arrayPeriod = np.linspace(1, 3, 4)
arrayIncl = np.linspace(70, 90, 4)
arraySma = np.linspace(5, 15, 3)
arrayQ = np.linspace(0.6, 0.8, 4)
arrayTeffP = np.linspace(7000, 7900, 3)
arrayTeffS = np.linspace(4500, 5500, 3)

for period in arrayPeriod:
    for incl in arrayIncl:
        for sma in arraySma:
            for q in arrayQ:
                for teffP in arrayTeffP:
                    for teffS in arrayTeffS:
                        print(count)
                        b = phoebe.default_binary()
                        b.add_constraint('semidetached', 'primary')
                        times = np.linspace(0, period, 100)
                        b.add_dataset('lc', times=times)
                        b["period@binary"] = period * u.day  # Orbital period in t
                        b["incl@binary@orbit@component"] = incl * u.deg
                        b["sma@binary@orbit@component"] = sma * u.solRad  # AU
                        b["q@binary@orbit@component"] = q  # mass ratio
                        b["teff@primary"] = teffP * u.K
                        b["teff@secondary"] = teffS * u.K

                        b.set_value('atm', component='primary', value='blackbody')
                        b.set_value('ld_mode', component='primary', value='manual')

                        try:
                            b.run_compute()
                            model = b["fluxes@latest@model"].get_value()
                            noise = np.random.randn(len(times)) / 100.0
                            model = model + noise
                            normflux_model = model / np.average(model)

                            fig, ax = plt.subplots(nrows=1, ncols=1)
                            ax.plot(times, normflux_model, "r-")
                            fig.savefig(cwd + "/Beta_Lyrae_Noisy/beta_lyrae" + str(count) + ".png")
                            plt.close(fig)

                            file2wr = open(cwd + "/Beta_Lyrae_Noisy/beta_lyrae" + str(count) + ".txt", "w")
                            # str2wr = "time\tflux\n"
                            # file2wr.write(str2wr)
                            for i, t in enumerate(times):
                                str2wr = "{:.6f}\t{:.4f}\n".format(t, normflux_model[i])
                                file2wr.write(str2wr)
                            file2wr.close()
                        except:
                            print("Şu parametrelerle sentetik B Lyr lc oluşturulamadı: count:" + str(
                                count) + " period:" + str(period) + " incl:" + str(incl) + " sma:" + str(
                                sma) + " q:" + str(q) + " teffP:" + str(
                                teffP) + " teffS:" + str(teffP))

                        count += 1
```
