```python
import os
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from google.colab import drive
import joblib

# Veri yolları
drive.mount('/content/gdrive')
path_Algol = '/content/gdrive/MyDrive/Doktora/Python/Algol_train'
path_Beta_Lyrae = '/content/gdrive/MyDrive/Doktora/Python/Beta_Lyrae_train'
path_W_UMa = '/content/gdrive/MyDrive/Doktora/Python/W_UMa_train'

# Tüm dosya yollarını al
Algol_files = [os.path.join(path_Algol, f) for f in os.listdir(path_Algol) if f.endswith('.txt')]
Beta_Lyrae_files = [os.path.join(path_Beta_Lyrae, f) for f in os.listdir(path_Beta_Lyrae) if f.endswith('.txt')]
W_UMa_files = [os.path.join(path_W_UMa, f) for f in os.listdir(path_W_UMa) if f.endswith('.txt')]

# Tüm dosyaların içeriğini birleştir
all_files = Algol_files + Beta_Lyrae_files + W_UMa_files

# Tüm dosyaların içeriğini okuyun ve verileri depolayın
data = []
labels = []

for file in all_files:
    with open(file, 'r') as f:
        lines = f.readlines()
        for line in lines:
            # Satırları ayrıştırın ve x, y koordinatlarını depolayın
            x, y = line.strip().split('\t')
            data.append([float(x), float(y)])

            # Sınıf etiketlerini depolayın
            if file in Algol_files:
                labels.append(0)
            elif file in Beta_Lyrae_files:
                labels.append(1)
            elif file in W_UMa_files:
                labels.append(2)

# Verileri ve etiketleri numpy dizilerine dönüştürün
data = np.array(data)
labels = np.array(labels)

# Verileri eğitim ve test kümelerine ayırın
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)

# Özellik ölçeklendirme
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Destek Vektör Makinesi (SVM) sınıflandırıcısını kullanarak modeli eğitin
model = SVC(kernel='rbf', gamma='scale', C=1.0)
model.fit(X_train, y_train)

# Test verileri üzerinde model doğruluğunu değerlendirin
accuracy = model.score(X_test, y_test)
print("Model accuracy: {:.2f}%".format(accuracy * 100))

#Modeli diske kaydet
joblib.dump(model, "/content/gdrive/MyDrive/Doktora/Python/SVM.joblib")
```
